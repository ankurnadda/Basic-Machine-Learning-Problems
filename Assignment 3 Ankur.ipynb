<h4> Problem #1-3</h4>  <br>
- Explain Bias-variance trade-off in choosing between LDA and QDA


With p predictors, estimating a co variance matrix requires estimating p(p+1)/2 parameters. The QDA estimates a separate co variance matrix for each class, so as the number of predictors becomes high, we experience a computational expense. Conversely, if we assume a common co variance matrix, we only have to do the computation once. LDA is a much less flexible classifier, than QDA, thus has substantially lower variance. However, if the assumption of uniform variance is highly off, then LDA can suffer high bias. In general, LDA tends to be better than QDA if there are relatively few training observations, so therefore reducing variance is crucial. QDA is recommended if the training set is very large, so that the variance of the classifier is not a major concern.

<b>Limitations of using Maximal Margin Classifier:</b>

Maximal margin classifier is a very natural way to perform classification, is a separating hyperplane exists. However existence of such a hyperplane may not be guaranteed, or even if it exists, the data is noisy so that maximal margin classifier provides a poor solution. If a new data point is added the decision boundary changes drastically

The maximum margin classifier depends on the support vectors and not on other training observations, hence hope is that largest margin on training data will also work well on test data.

Assumes that the classes can be perfectly separable, which is not always the case.


<b>Difference between SVC and MMC:</b>

The generalization of the maximal margin classifier to the non-separable case is known as the support vector classifier, where a small proportion of the training sample is allowed to cross the margins, or even the separating hyperplane. 

Rather than looking for the largest possible margin so that every observation is on the correct side of the margin, thereby making the margins very narrow or non-existent, some observations are allowed to be on the incorrect side of the margins. The margin is soft as a small number of observations violate the margin. The softness is controlled by slack variables which control the position of the observations relative to the margins and separating hyperplane. 

<b> Impact of Slack Variables </b>

Large:
‚Äì Allow for more violation of margin
‚Äì More support vectors
‚Äì Less variance, more stable
‚Äì Potential high bais

Small:
‚Äì less violation on training data allowed
‚Äì Low training training error, less bias
‚Äì Fewer support vectors
‚Äì Higher variance

<h4> Problem #3-3 Cross validation, SVM </h4> <br>
In this exercise, we will use SVM  for breast cancer classification (malignant or benign).<br> 
Use a code below to download the dataset:


```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
dataset = load_breast_cancer()
print(dataset.keys())
```

Follow steps to answer questions.
- Scale the features with MinMaxScaler
- Split breast cancer dataset into two, test dataset and train dataset
- Find best SVM classifier model. Try values of C=[0.001, 0.1, 1, 10, 1000] and Gamma = [0.001, 0.1, 1, 10, 1000]. 
- Use 3-fold cross validation to find the best parameters (using all possible combinations of these values for C and gamma).
- Report your parameters
- Report accuracy, confusion matrix, precision, and recall (use Test dataset, SVM classifier model with best parameters)
- Comment your results

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix, precision_score, recall_score
import itertools
import matplotlib.pyplot as plt
import numpy as np

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')

kfolds = 3
c = [0.001, 0.1, 1, 10, 1000]
g = [0.001, 0.1, 1, 10, 1000]
bestScore = 0
data = dataset.data
scaler = MinMaxScaler()
scaler.fit(data)
scaler.data_max_
scaled_data = scaler.transform(data)
X_train, X_test, Y_train, Y_test = train_test_split(scaled_data, dataset.target, random_state=0)
for i in range(0, 5):
    svmModel = SVC(kernel='rbf', gamma=g[i], C=c[i])
    scores = cross_val_score(svmModel, X_train, Y_train, cv=kfolds)
    score = np.mean(scores)
    if score > bestScore:
        bestScore = score
        bestParameter = c[i]
        bestGama = g[i]
FinalsvmModel = SVC(kernel='rbf', gamma=bestGama, C=bestParameter).fit(X_train, Y_train)
test_score = FinalsvmModel.score(X_test,Y_test)
PredictedOutput = FinalsvmModel.predict(X_test) 
print("Best Score on train set", bestScore)
print("Best Parameter ", bestParameter)
print("Best Gamma ", bestGama)
print("Score of SVM ", test_score)
confusion=confusion_matrix(Y_test,PredictedOutput)
plt.figure()
plot_confusion_matrix(confusion, classes=dataset.target_names,title='Confusion matrix')
plt.show()
print("Precision ", precision_score(Y_test,PredictedOutput))
print("Recall ", recall_score(Y_test,PredictedOutput))
print("-------------------------")    

```

Cross validation gives us the optimal parameters and gamma for the SVM model which reflects best scores
