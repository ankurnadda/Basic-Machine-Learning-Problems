The table below provides a training data set containing six observations, three predictors, and one qualitative response variable.
Suppose we wish to use this data set to make a prediction for Y when X1 = X2 = X3 = 0 using K-nearest neighbors.

|Obs.|X1|X2|X3|Y
|--|-------------------------------|
|1 |0|3|0|Green|
|2 |2|0|0|Green|
|3|0|1|3|Green|
|4|0|1|2|Red|
|5|-1|0|1|Red|
|6|1|1|1|Green|

<b>Problem #1-1.</b> Compute the Euclidean distance between each observation and the test point, X1 = X2 = X3 = 0

```python
#Write your narrative answer here
import numpy as np
from scipy.spatial import distance
import warnings
warnings.filterwarnings("ignore")
input_arr = np.array([[0,3,0],[2,0,0],[0,1,3],[0,1,2],[-1,0,1],[1,1,1]])  # Generate Numpy Array
test_case = np.array([0,0,0])
dst=[]
for i in range(0,6):
    temp = distance.euclidean(test_case,input_arr[i])  # Calculate euclidean distance for each observation and test case
    dst.append(temp)                                   # Store distances in a list for future use.
    print("The euclidean distance between test case and observation ",i+1,"is",temp)
```

<b>Problem #1-2.</b> What is our prediction with K = 1? and explain why.

```python
#Write your narrative answer here
# when K = 1, select min value from the list of distances calculated
index_min = np.argmin(dst)
print("The minimum distance and corrosponding observation index are printed below")
print(dst[index_min])
print(index_min+1)
```

The observation 5 corrosponds to Red in the training data, hence the resulting test case belongs to Red. When K =1, select the nearest or min distance value.

<b>Problem #1-3.</b> What is our prediction with K = 3? and explain why.

```python
#Write your narrative answer here
sort_index = np.argsort(dst)
print("The observations sorted on the basis of euclidean distance from min to max")
print(sort_index+1)
```

For K =3, pick 3 nearest neighbors or top 3 min values from the Euclidean distance list. The top 3 minimum observations are 5(red), 6(green) and 2(green). When K =3, the test case will be classified as Green instead of red in case of k=1. The reason for the same is that probability of test case as Green is 2/3 as compared to 1/3 for red.

Choice of K has large impact on the performance. Large values of K are suitable for large dataset (KNN in general is suitable for slightly large dataset as compared to above example). <b>Problem #4-1.</b>  Ordinary Least Square (OLS), calculate parameters

Follow two steps<br> 

- From your DataFrame generate new DataFrame with 2 columns ('LSTAT' and target)  

- Using the least squared formula below write a function to calulate $B_0$  and $B_1$  from your new DataFrame where 'LSTAT' = x and target = y. *(hint you can add 2 colums to your DataFrame($x*x$  and  $x*y$), check the problem #2-1)*

\begin{array} \\
    \hat{\beta}_0 = \bar{y} - \hat{\beta}_1\bar{x}, \\ \\
    \hat{\beta}_1 = \frac{\sum_{i=1}^n y_i x_i - \bar{y}\sum_{i=1}^n x_i}{\sum_{i=1}^n x^2 - \bar{x}\sum_{i=1}^n x_i} \\ \\
    \text{where }\bar{x} = \frac{\sum_{i=1}^y x_i}{n} \text{ and } \bar{y} = \frac{\sum_{i=1}^y y_i}{n} \\
\end{array}

```python
#Write your narrative answer here
df_2= pd.DataFrame()
df_2["x"] = df["LSTAT"]
df_2["y"] = df["Target"]
x_mean = df_2["x"].mean()
y_mean = df_2["y"].mean()
y1 = []
x1=[]
x1y1=[]
x1_2 =[]
for index, row in df_2.iterrows():
    temp= (row["x"]-x_mean)
    temp1= (row["y"]-y_mean)
    temp2=(temp*temp1)
    temp3=temp*temp
    x1.append(temp)
    y1.append(temp1)
    x1y1.append(temp2)
    x1_2.append(temp3)
df_2["x1"] = x1
df_2["y1"] = y1
df_2["x1y1"] = x1y1
df_2["x1^2"]=x1_2
b1= (df_2["x1y1"].sum()/df_2["x1^2"].sum())
b0=y_mean - (b1*x_mean)
print("Coef ",b1)
print("Coef ", b0)
```

\begin{array} \\
