1. Create a dictionary of score of each team. That is keys:values are
Team1: 4,
Team2: 3,
Team3: 5,
Team4: 2

2. Add to the dictionary ‘Team5’ who has a score ‘5’ 

3. PRINT all the keys of the dictionary using the keys() method

4. Find the length VALUES in the dictionary (use the len function) and print it

5. Get the average score of the teams and print it (use the sum and len functions)


```python
Score = {'Team1': 4,'Team2': 3, 'Team3': 5, 'Team4': 2} # Dictionary of score for 4 team
Score['Team5']=5  # Addition of new team 5
print(Score.keys())  # Print all the keys
print(Score.values()) # Print values
l=len(Score) # Calulate length
print(l) # Print lenght of dict
Average= sum(Score.values())/l # Calculate Average
print(Average) # Print Average score
```

B) 
1. Generate a numpy array with 2 columns, where first column contains numbers from 0 to 5 and second column is [0, 1, 4, 9, 16, 25]

2. Check the shape of your array. It should be 6x2.

3. Put the array into a data frame, with column labels ‘x’ and ‘y’ 

4. Plot ‘x’ versus ‘y’ 

5. Get average of elements in second column of the data frame


```python
%matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

arr = np.array([[0,0],[1,1],[2,4],[3,9],[4,16],[5,25]])  # Generate Numpy Array
print(arr.shape)  # Check shape of the array
df = pd.DataFrame(arr) # Put array to dataframe 
df.columns = ['x', 'y'] # Add Column labels to dataframe
print(df) # print dataframe
plt.scatter(df['x'], df['y']) # plot points, x versus y
plt.show()
Avg = df['y'].sum()/df['y'].count() # Calculate average of second colunm y
print(Avg) # Print Average
```
