<h3> Problem #2. Feature Selection  [10 points] </h3>  <br>
In this part, you should download and analyze **"Boston House Prices" ** dataset **with only 4 features**. <br>
(use code below to download  and change dataset)

- Find the best two features using forward and backward selection to the Boston House Prices" dataset with only 4 features ('CHAS','RM','TAX', 'LSTAT'), use a linear regression model to find a best subset of features.
- Comment your choice 

```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_boston
import statsmodels.formula.api as smf
dataset = load_boston()
import numpy as np 
import pandas as pd
#print(dataset.keys())
# Data preprocessing
data_1= pd.DataFrame(dataset.data, columns=dataset.feature_names)[['CHAS','RM','TAX', 'LSTAT']]
data_1['target'] = dataset.target
#X_train,X_test,Y_train,Y_test=train_test_split(new_data.values,dataset.target,random_state=0)
```

```python
#Write your narrative answer here
#forward Selection
import sys
features = data_1.columns.values
p_score_first = []
p_score_second = []
selected = []
maximum = sys.maxsize
minimum = -sys.maxsize
for i in range(0, 4):
    model = smf.ols('target~{}'.format(features[i]), data_1)
    result = model.fit()
    p_score_first.append(result.pvalues[1])
skip = p_score_first.index(min(p_score_first))
selected.append(features[p_score_first.index(min(p_score_first))])
for i in range(0, 4):
    if i != skip:
        model = smf.ols('target~{}+{}'.format(selected[0], features[i]), data_1)
        result = model.fit()
        p_score_second.append(result.pvalues[1])
    else:
        p_score_second.append(maximum)
selected.append(features[p_score_second.index(min(p_score_second))])
print("The best features for Forward Selection are ", selected)

#Backward Selection
features = data_1.columns.values
p_score_first = []
p_score_second = []
selected = []
for i in range(0, 4):
    model = smf.ols('target~{}'.format(features[i]), data_1)
    result = model.fit()
    p_score_first.append(result.pvalues[1])
skip = p_score_first.index(max(p_score_first))
selected.append(features[p_score_first.index(max(p_score_first))])
for i in range(0, 4):
    if i != skip:
        model = smf.ols('target~{}+{}'.format(selected[0], features[i]), data_1)
        result = model.fit()
        p_score_second.append(result.pvalues[1])
    else:
        p_score_second.append(minimum)
selected.append(features[p_score_second.index(max(p_score_second))])
print("The best features for Backward Selection are ", selected)
```

The forward and backward selection will yield different results as the approach for the same is based on p-values in case of the above defined function. In forward selection, the feature are selected with respect to the initial feature selected and added or rejected accordingly. Backward selection is opposite to forward in a way that, all features are intially selected and then removed each by each based on p values.
